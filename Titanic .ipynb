# Filling Age column by considering their family count

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1  

df['Age'] = df.groupby(['FamilySize', 'Survived'])['Age'].transform(lambda x: x.fillna(x.median()))

df['Age'] = df.groupby(['FamilySize'])['Age'].transform(lambda x: x.fillna(x.median()))

df['Age'].fillna(df['Age'].median(), inplace=True)

df[['FamilySize', 'Survived', 'Age']]
```

```python
df
```

# If 2 or more people have same ticket and someone does not have embarked value then match and assign the values

```python
null_embarked = df[df['Embarked'].isnull()]

for index, row in null_embarked.iterrows():
    ticket = row['Ticket']
    
    # Find rows with the same ticket and non-null 'Embarked'
    same_ticket = df[(df['Ticket'] == ticket) & (df['Embarked'].notnull())]
    
    if not same_ticket.empty:
        # Assign the 'Embarked' value from the first matching row
        df.loc[index, 'Embarked'] = same_ticket['Embarked'].iloc[0]
        
```

```python
most_common_embark = df['Embarked'].mode()[0]
df['Embarked'].fillna(most_common_embark, inplace=True)

df.info()
```

# Encoding Categorical Data

```python
df = pd.get_dummies(df)
df
```

```python
df["Survived"].value_counts()
```

```python
X = df.drop("Survived",axis=1)
y = df["Survived"]
```

```python
from imblearn.over_sampling import SMOTE

X_res, y_res = SMOTE().fit_resample(X,y)
y_res.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_res,y_res,test_size = 0.20, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train
```

# Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score, recall_score, f1_score

rf = RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```

```python
f1_score(y_test,y_pred)
```
